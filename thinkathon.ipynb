```python
# Perform data cleaning
df_cleaned = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# Display the cleaned DataFrame
print("\nCleaned DataFrame:")
print(df_cleaned)
```

```python
# Perform data reduction (drop unnecessary columns)
columns_to_drop = ['img_link', 'product_link']
df_reduced = df_cleaned.drop(columns=columns_to_drop)

# Display the reduced DataFrame
print("\nReduced DataFrame:")
print(df_reduced)
```

```python
# Perform data transformation (convert string columns to numeric)
df_transformed = df_reduced.copy()
df_transformed['discounted_price'] = pd.to_numeric(df_transformed['discounted_price'], errors='coerce')

# Display the transformed DataFrame
print("\nTransformed DataFrame:")
print(df_transformed)
```

```python
# Remove duplicate values
df_no_duplicates = df_transformed.drop_duplicates()

# Display the DataFrame after removing duplicates
print("\nDataFrame after removing duplicates:")
print(df_no_duplicates)
```
